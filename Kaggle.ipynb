```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

# Parte 1 - Leitura do dataset

Esse notebook realiza a análise exploratória do [dataset] (https://www.kaggle.com/dataset/e626783d4672f182e7870b1bbe75fae66bdfb232289da0a61f08c2ceb01cab01) disponibilizado pelo Hospital Israelita Albert Einstein (SP), contendo 5644 amostras e 111 atributos, com o objetivo de realizar a predição de diagnóstico do Vírus Covid-19

```python
data = pd.read_excel('data/dataset.xlsx')  
```

```python
data.columns.values ### visualização dos atributos
```

# Parte 2 - Remover colunas

Foram removidas 5 colunas do dataset:
1. Patient ID: nome do paciente
2. Mycoplasma pneumoniae, Urine - Sugar,
  Partial thromboplastin time (PTT), 
  Prothrombin time (PT), 
  Activity,
  D-Dimer: Esses atributos possuem a coluna completamente vazia e foram removidos 

```python
data.columns[np.where(data.isnull().sum().values==5644)[0]]
```

```python
data = data.drop('Patient ID',axis=1)
```

```python
data = data.drop(data.columns[np.where(data.isnull().sum().values==5644)[0]],axis=1)
```

# Parte 3 - Análise de valores NaN 

Além das colunas removidas por estarem com todos os valores NaN, verificou-se também a quantidade de valores vazios na base. Desse total, apenas 4 atributos não possuem valores NaN, enquanto os demais a maioria possui mais que a metade de valores vazios

```python
total = np.full((105, 1), 5644)
```

```python
data.isnull().sum().values
```

```python
f, ax = plt.subplots(figsize=(20, 20))
sns.set_color_codes("pastel")
sns.barplot(x=total.T[0], y=data.columns,color="b")
sns.set_color_codes("muted")
sns.barplot(x=data.isnull().sum().values, y=data.columns,color='b')
```
